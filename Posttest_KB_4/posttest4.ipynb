## Import Library

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv("games.csv")

data.info()
```

```python
def drop_columns(df: pd.DataFrame, cols_to_drop: list = None, 
                 idx_to_drop: list = None) -> pd.DataFrame:
    new_df = df.copy()

    if cols_to_drop is not None:
        new_df = new_df.drop(columns=cols_to_drop)

    if idx_to_drop is not None:
        new_df = new_df.drop(index=idx_to_drop)

    return new_df


def convert_str_num_to_int(df: pd.DataFrame, col_names: list) -> pd.DataFrame:
    new_df = df.copy()
    
    for col in col_names:
        new_df[col] = new_df[col].str.replace("K", "000")

        new_df[col] = new_df[col].map(lambda x: int(float(x) * 1000) if '.' in x else int(x))

    return new_df
```

```python
numerical_cols = ["Plays", "Playing", "Backlogs", "Wishlist"]

data = convert_str_num_to_int(data, numerical_cols)
```

```python
data = drop_columns(data, "Unnamed: 0")
```

```python
data.info()
```

## Membagi Dataset

```python
X = data.iloc[:, : -1]
y = data.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

print("X_train", X_train.shape)
print("X_test", X_test.shape)
print("y_train", y_train.shape)
print("y_test", y_test.shape)
```

## Melakukan Normalisasi

```python
min_max = MinMaxScaler()

# Copy Data
copy1 = data
dataCopy_normalization = copy1[["Playing"]]

normalization = min_max.fit_transform(dataCopy_normalization[["Playing"]])

data_normalization = pd.DataFrame(normalization)
data_normalization.columns =["Playing_Normalization"]

dataCopy_normalization = dataCopy_normalization.join(data_normalization["Playing_Normalization"]).reset_index()
dataCopy_normalization = dataCopy_normalization.drop("index", axis=1)

print()
print("Data Sebelum dan Sesudah dilakukan normalisasi : \n")
dataCopy_normalization.head(10)
```

## Data Cleaning Nilai Null
